```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("laliga_df.csv")
df["count"]=1
df.head()
```

```python
df['localTeam_won']=df.apply(lambda row: 1 if row['localGoals']>row['visitorGoals'] else 0,axis=1)
df['visitorTeam_won']=df.apply(lambda row: 1 if row['localGoals']<row['visitorGoals'] else 0,axis=1)
df['Draw']=df.apply(lambda row: 1 if row['localGoals']==row['visitorGoals'] else 0,axis=1)
```

```python
home_table=df
home_table["Home_Losses"] =0
home_table=home_table.drop(["visitorTeam","date","id","round"],axis=1)
home_table=home_table.rename(columns={"localTeam":"Club"})
```

```python
home_table=home_table.groupby(["season","Club"])[["count","localTeam_won","Draw","Home_Losses","localGoals","visitorGoals"]].sum()
```

```python
home_table=home_table.rename(columns={"count":"Home_Matches_Played","localTeam_won":"Home_Wins",
                             "localGoals":"Home_Goals_Scored","Draw":"Home_Draws",
                                     "visitorGoals":"Home_Goals_Concede"})
```

```python
home_table["Home_Losses"]=home_table["Home_Matches_Played"]-home_table["Home_Wins"]-home_table["Home_Draws"]
```

```python
home_table["Home_Points"]=(3*home_table["Home_Wins"])+home_table["Home_Draws"]
```

```python
home_table.head()
```

```python
away_table=df
away_table["Away_Losses"]=0
away_table=away_table.drop(["localTeam","date","id","round"],axis=1)
away_table=away_table.rename(columns={"visitorTeam":"Club"})
```

```python
away_table=away_table.groupby(["season","Club"])[["count","visitorTeam_won","Draw","Away_Losses","visitorGoals","localGoals"]].sum()
```

```python
away_table=away_table.rename(columns={"count":"Away_Matches_Played","visitorTeam_won":"Away_Wins",
                             "visitorGoals":"Away_Goals_Scored","Draw":"Away_Draws",
                                     "localGoals":"Away_Goals_Concede"})
```

```python
away_table["Away_Losses"]=away_table["Away_Matches_Played"]-away_table["Away_Wins"]-away_table["Away_Draws"]
```

```python
away_table["Away_Points"]=(3*away_table["Away_Wins"])+away_table["Away_Draws"]
```

```python
away_table.head()
```

```python
stat_table=pd.concat([home_table,away_table],axis=1)
```

```python
stat_table
```

```python
point_table=stat_table
```

```python
point_table["Matches_Played"]=point_table["Home_Matches_Played"]+point_table["Away_Matches_Played"]
point_table=point_table.drop(["Home_Matches_Played","Away_Matches_Played"],axis=1)
```

```python
point_table["Wins"]=point_table["Home_Wins"]+point_table["Away_Wins"]
point_table=point_table.drop(["Home_Wins","Away_Wins"],axis=1)
```

```python
point_table["Draws"]=point_table["Home_Draws"]+point_table["Away_Draws"]
point_table=point_table.drop(["Home_Draws","Away_Draws"],axis=1)
```

```python
point_table["Losses"]=point_table["Home_Losses"]+point_table["Away_Losses"]
point_table=point_table.drop(["Home_Losses","Away_Losses"],axis=1)
```

```python
point_table["Goals_Scored"]=point_table["Home_Goals_Scored"]+point_table["Away_Goals_Scored"]
point_table=point_table.drop(["Home_Goals_Scored","Away_Goals_Scored"],axis=1)
```

```python
point_table["Goals_Concede"]=point_table["Home_Goals_Concede"]+point_table["Away_Goals_Concede"]
point_table=point_table.drop(["Home_Goals_Concede","Away_Goals_Concede"],axis=1)
```

```python
point_table["Goal_Diffrence"]=point_table["Goals_Scored"]-point_table["Goals_Concede"]
```

```python
point_table["Points"]=point_table["Home_Points"]+point_table["Away_Points"]
point_table=point_table.drop(["Home_Points","Away_Points"],axis=1)
```

```python
point_table.head()
```

```python
point_table=point_table.sort_values(["season","Points"], ascending=[True,False])
```

```python
point_table
```

```python
season=point_table.index.levels[0]
```

```python
Club=[]
Points=[]
Year=[]

for i in season:
    Club.append((point_table.loc[i]).index[0])
    Club.append((point_table.loc[i]).index[1])
    Points.append(((point_table.loc[i,"Points"])[0]))
    Points.append(((point_table.loc[i,"Points"])[1]))
    Year.append(i)
    Year.append(i)

Season_Winner_temp=pd.DataFrame({"Season":Year,"Club":Club,"Points":Points})
```

```python
Season_Winner_temp
```

```python
champ=pd.DataFrame({'Season':['a'],'Club':['a'],'Points':['1']})
```

```python
df=pd.read_csv("laliga_df.csv")
for i in season:
    df1=Season_Winner_temp.loc[Season_Winner_temp['Season']==i].reset_index()
    if (((df1['Points'])[0])!=((df1['Points'])[1])):
        champ=pd.concat([champ,pd.DataFrame(df1.iloc[0:1,:])])
    else:
        df2=df.loc[df["season"]=="i"]
        team1=(df1['Club'])[0]
        team2=(df1['Club'])[1]
        df3=df2.loc[((df2["localTeam"]==team1)|(df2["localTeam"]==team2))&((df2["visitorTeam"]==team1)|(df2["visitorTeam"]==team2))].reset_index()
        
        team1goal,team2goal = 0,0
        
        for j in range(df3.shape[0]):
            if (df3.iloc[j,4]==team1):
                team1goal = df3["localGoals"][i]
            else:
                team2goal = df3["localGoals"][i]
            if (df3.iloc[j,5]==team1):
                team1goal+=df3["visitorGoals"][i]
            else:
                team2goal+=df3["visitorGoals"][i]
        if(team1goal>team2goal):
            champ=pd.concat([champ,pd.DataFrame(df1.iloc[0:1,:])])
        else:
            champ=pd.concat([champ,pd.DataFrame(df1.iloc[1:2,:])])

            
```

```python
champ.set_index('Season',inplace=True)
champ=champ.drop('a')
champ=champ.drop('index',axis=1)
```
